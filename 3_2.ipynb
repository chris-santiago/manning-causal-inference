```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FILE = 'diffs.parq.gzip'

sns.set_style("darkgrid")
pd.options.display.max_columns = 100
```

```python
data = pd.read_parquet(FILE)
data.head()
```

```python
data['post_intervention'] = data['month'] >= 5
data['treated'] = data['gender'] == 'W'
data.head()
```

```python
data.groupby(['post_intervention', 'treated'])['discount'].mean()
```
