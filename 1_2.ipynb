```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
FILE = 'clean.parq.gzip'

sns.set_style("darkgrid")
```

```python
def get_net_revenue(sales, margin=.05):
    return sales * margin


def get_profit(sales, discount, **kwargs):
    return get_net_revenue(sales, **kwargs) - discount
```

```python
data = pd.read_parquet(FILE)
data.head()
```

```python
sns.regplot(
    x='discount', 
    y='sales', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```

```python
data['profit'] = get_profit(data['sales'], data['discount'])
```

```python
sns.regplot(
    x='discount', 
    y='profit', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```

```python
sns.regplot(
    x='age', 
    y='discount', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```

```python
sns.regplot(
    x='sales_prediction', 
    y='discount', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```

```python
sns.boxplot(
    x='cust_state', 
    y='discount', 
    data=data,
)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.regplot(
    x='age', 
    y='discount', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```

```python
sns.regplot(
    x='spend_1m_groceries', 
    y='discount', 
    data=data,
    lowess=True,
    line_kws={'color': 'orange'}
)
plt.show()
```
